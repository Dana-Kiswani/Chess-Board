```python
import numpy as np
import matplotlib.pyplot as plt

class ChessBoard:
    white = [1,1,1]
    blue = [0,0,1]
    red = [1,0,0]
    
    def __init__(self):
        self.red_queen=None
        self.blue_queen=None
        self.under_attack=False
        self.grid = np.zeros([8,8,3])
        for i in range(8):
            for j in range(8):
                if (i%2==0 and j%2==0) or (i%2==1 and j%2==1):
                    self.grid[i,j] = ChessBoard.white
    
    def render(self):
        plt.imshow(self.grid)
    
    
    def is_under_attack(self):
        pass
    
    def add_red(self, x, y):
            #row->x / colums->y
            # self.row = x
            # self.colum = y
            self.red_queen=(x,y)
            self.grid[x,y] = ChessBoard.red
            
    def add_blue(self, d, b):
            #row->x / colums->y
            # self.row = d
            # self.colum = b
            self.blue_queen=(d,b)
            self.grid[d,b] = ChessBoard.blue
            
    def check_if_under_attack(self):

        if self.blue_queen[0]==self.red_queen[0]:
            self.under_attack=True
            print('under attack')
            return self.under_attack

        elif self.blue_queen[1]==self.red_queen[1]:
            self.under_attack=True
            print('under attack')
            return self.under_attack

        elif np.absolute(self.blue_queen[0]-self.red_queen[0])==np.absolute(self.blue_queen[1]-self.red_queen[1]):
            self.under_attack=True
            print('under attack')
            return self.under_attack

        else:
            self.under_attack = False
            return self.under_attack

board = ChessBoard()

board.render()
```

```python
#Calling the red queen..

board = ChessBoard()
board.add_red(6,6)
board.add_blue(6,0)
board.check_if_under_attack()
board.render()
```

```python
#Calling the blue queen..

board = ChessBoard()
board.add_blue(2,6)
board.add_red(3,6)
board.check_if_under_attack()
board.render()
```

```python
# Both of them ..

board = ChessBoard()
board.add_red(6,6)
# board.render()
board.add_blue(4,4)
board.check_if_under_attack()
board.render()
```

```python
# Tests :

#1)queens on same row should be “under attack”
board.add_red(6,6)
# board.render()
board.add_blue(6,4)
assert board.check_if_under_attack()==True
print('same row --> PASSED')

#2)queens on same column should be “under attack”
board.add_red(6,6)
# board.render()
board.add_blue(4,6)
assert board.check_if_under_attack()==True
print('same column --> PASSED')

#3)queens on same diagonal should be “under attack”
board.add_red(6,6)

board.add_blue(4,4)
assert board.check_if_under_attack()==True
print('same diagonal --> PASSED')

#4)queens with any other coordinates should NOT be “under attack”
board.add_red(1,5)
# board.render()
board.add_blue(7,6)
assert board.check_if_under_attack()==False
print('Is not Under attack --> PASSED')
```
